```python
#Splitting multi-genre entries and counting individual genres
genre_counts = {}
for genres in df['genre']:
    # Split by comma and clean whitespace
    genre_list = [g.strip() for g in genres.split(',')]
    for genre in genre_list:
        genre_counts[genre] = genre_counts.get(genre, 0) + 1
        
genre_df = pd.DataFrame(list(genre_counts.items()), columns=['Genre', 'Count'])
genre_df = genre_df.sort_values('Count', ascending=False)
```

```python
# Visualization of the Bar chart of genre distribution
plt.figure(figsize=(14, 6))
plt.bar(genre_df['Genre'], genre_df['Count'], color='steelblue', edgecolor='black')
plt.xlabel('Genre', fontsize=12, fontweight='bold')
plt.ylabel('Number of Songs', fontsize=12, fontweight='bold')
plt.title('Distribution of Songs Across Music Genres', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# The genre that has the most sonngs is:
print("Which genre has the most songs?")
top_genre = genre_df.iloc[0]
print(f"\n Genre with most songs: {top_genre['Genre']} ({top_genre['Count']} songs)")
```

```python
# The percentage of songs in top 3 genres are:
top_3_genres = genre_df.head(3)
total_genre_tags = genre_df['Count'].sum()
top_3_percentage = (top_3_genres['Count'].sum() / total_genre_tags) * 100

print("What percentage of songs are in the top 3 genres?")
print(f"\n Top 3 Genres:")
for idx, row in top_3_genres.iterrows():
    percentage = (row['Count'] / total_genre_tags) * 100
    print(f"   {row['Genre']}: {row['Count']} songs ({percentage:.1f}%)")
print(f"\n   Combined: {top_3_percentage:.1f}% of all genre tags")
```

```python
#Create correlation heatmap
# Select relevant numerical features for correlation analysis
audio_features = ['popularity', 'energy', 'danceability', 'loudness', 
                  'valence', 'tempo', 'acousticness', 'speechiness', 
                  'instrumentalness', 'liveness', 'duration_ms']
```

```python
# Calculating the correlation matrix
correlation_matrix = df[audio_features].corr()
```

```python
# Creating the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Song Features', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
# Identifying the strongest correlations
print("\nStrongest Positive Correlations:")

upper_triangle = correlation_matrix.where(           # obtaining the upper triangle of correlation matrix and avoid duplicates
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
)
correlations_sorted = upper_triangle.stack().sort_values(ascending=False)  # Stacking and sorting correlations


# Displaying the top 5 positive correlations
for i, (features, corr_value) in enumerate(correlations_sorted.head(5).items(), 1):
    print(f"   {i}. {features[0]} ↔ {features[1]}: {corr_value:.3f}")

print("\n Strongest Negative Correlations:")
# Display top 5 negative correlations
for i, (features, corr_value) in enumerate(correlations_sorted.tail(5).items(), 1):
    print(f"   {i}. {features[0]} ↔ {features[1]}: {corr_value:.3f}")
```

```python
#Analyze surprising correlations
print("\n Key Insights:")

# Energy vs Acousticness (usually negative)
energy_acoustic_corr = correlation_matrix.loc['energy', 'acousticness']
print(f"\n   Energy ↔ Acousticness: {energy_acoustic_corr:.3f}")
if energy_acoustic_corr < -0.5:
    print("   → Strong negative relationship: High-energy songs tend to be less acoustic")
    print("      (Makes sense: electric/loud instruments create energy)")

# Energy vs Loudness (usually positive)
energy_loudness_corr = correlation_matrix.loc['energy', 'loudness']
print(f"\n   Energy ↔ Loudness: {energy_loudness_corr:.3f}")
if energy_loudness_corr > 0.5:
    print("   → Strong positive relationship: Energetic songs are louder")
    print("      (Expected: Rock/EDM are both energetic and loud)")

# Valence vs Energy
valence_energy_corr = correlation_matrix.loc['valence', 'energy']
print(f"\n   Valence ↔ Energy: {valence_energy_corr:.3f}")
if valence_energy_corr > 0.3:
    print("   → Moderate positive relationship: Happy songs tend to be energetic")
    print("      (Makes sense: Upbeat = energetic + positive)")
elif valence_energy_corr < 0:
    print("   → Surprising! Negative relationship between happiness and energy")

# Popularity correlations
print(f"\n Popularity Correlations:")
pop_corrs = correlation_matrix['popularity'].drop('popularity').sort_values(ascending=False)
for feature, corr_val in pop_corrs.head(3).items():
    print(f"      {feature}: {corr_val:.3f}")
    
if abs(pop_corrs.max()) < 0.3:
    print("\n   → Surprising: NO strong predictors of popularity!")
    print("      This suggests popularity depends on factors beyond audio features")
    print("      (e.g., marketing, artist fame, cultural trends)")

```

**Task 2**

```python
# Get top 10 most popular songs
top_10_songs = df.nlargest(10, 'popularity')

#Creating a radar chart for top 10 songs
# Selecting features to display (normalizing between 0-1 for fair comparison)
radar_features = ['danceability', 'energy', 'valence', 'acousticness', 'speechiness']

# Preparing data for radar chart
fig, axes = plt.subplots(2, 5, figsize=(20, 8), subplot_kw=dict(projection='polar'))
axes = axes.flatten()

for idx, (song_idx, song) in enumerate(top_10_songs.iterrows()):
    ax = axes[idx]
    
    # Get feature values
    values = song[radar_features].values.tolist()
    values += values[:1]  # Close the circle
    
    # Create angles for radar chart
    angles = np.linspace(0, 2 * np.pi, len(radar_features), endpoint=False).tolist()
    angles += angles[:1]
    
    # Plot
    ax.plot(angles, values, 'o-', linewidth=2, color='darkblue')
    ax.fill(angles, values, alpha=0.25, color='skyblue')
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(radar_features, size=8)
    ax.set_ylim(0, 1)
    ax.set_title(f"{song['artist']}\n{song['song'][:20]}...", 
                 size=9, fontweight='bold', pad=10)
    ax.grid(True)

plt.suptitle('Audio Feature Profiles of Top 10 Most Popular Songs', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
    
```

```python
print("\n Top 10 Songs - Feature Analysis:")
print(f"\n{'Rank':<5} {'Artist':<20} {'Song':<30} {'Pop':<5} {'Dance':<6} {'Energy':<7} {'Valence':<8}")
print("-" * 85)

for rank, (idx, song) in enumerate(top_10_songs.iterrows(), 1):
    artist = song['artist'][:18]
    song_name = song['song'][:28]
    pop = song['popularity']
    dance = song['danceability']
    energy = song['energy']
    valence = song['valence']
    
    print(f"{rank:<5} {artist:<20} {song_name:<30} {pop:<5} {dance:<6.3f} {energy:<7.3f} {valence:<8.3f}")
```

```python
# Comparing the averages
print(f"\n{'Average':<5} {'':<20} {'':<30} {'':<5} {top_10_songs['danceability'].mean():<6.3f} {top_10_songs['energy'].mean():<7.3f} {top_10_songs['valence'].mean():<8.3f}")
print(f"{'Dataset Avg':<5} {'':<20} {'':<30} {'':<5} {df['danceability'].mean():<6.3f} {df['energy'].mean():<7.3f} {df['valence'].mean():<8.3f}")
```

```python
# Additional visualizations: Parallel coordinates plot
print("\n Creating Parallel Coordinates Visualization...")

# Normalize features to 0-1 scale for better comparison
features_to_plot = ['danceability', 'energy', 'valence', 'acousticness', 'loudness']
top_10_normalized = top_10_songs.copy()

# Normalize loudness (it's negative, so special handling)
top_10_normalized['loudness'] = (top_10_normalized['loudness'] - df['loudness'].min()) / (df['loudness'].max() - df['loudness'].min())

# Create parallel coordinates plot
fig, ax = plt.subplots(figsize=(14, 7))

# Create x-axis positions
x = np.arange(len(features_to_plot))

# Plot each song as a line
colors = plt.cm.rainbow(np.linspace(0, 1, len(top_10_normalized)))

for (idx, song), color in zip(top_10_normalized.iterrows(), colors):
    values = [song[feature] for feature in features_to_plot]
    ax.plot(x, values, marker='o', label=f"{song['artist'][:15]} - {song['song'][:15]}", 
            linewidth=2, markersize=8, color=color, alpha=0.7)

# Styling
ax.set_xticks(x)
ax.set_xticklabels(features_to_plot, rotation=15, ha='right')
ax.set_ylabel('Normalized Value (0-1)', fontsize=12, fontweight='bold')
ax.set_title('Parallel Coordinates: Top 10 Songs Feature Comparison', 
             fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
ax.set_ylim(-0.1, 1.1)

plt.tight_layout()
plt.show()
```

```python
# Analyze patterns
print("\n Feature Profile Insights:")

high_dance = top_10_songs['danceability'].mean()
high_energy = top_10_songs['energy'].mean()
high_valence = top_10_songs['valence'].mean()

print(f"\n   Top songs tend to have:")
if high_dance > df['danceability'].mean():
    print(f"   ✓ Higher danceability ({high_dance:.2f} vs {df['danceability'].mean():.2f} average)")
if high_energy > df['energy'].mean():
    print(f"   ✓ Higher energy ({high_energy:.2f} vs {df['energy'].mean():.2f} average)")
if high_valence > df['valence'].mean():
    print(f"   ✓ Higher valence/positivity ({high_valence:.2f} vs {df['valence'].mean():.2f} average)")

# Check for diversity in top songs
std_dance = top_10_songs['danceability'].std()
std_energy = top_10_songs['energy'].std()

print(f"\n   Diversity among top songs:")
if std_dance > 0.15:
    print(f"   • High variation in danceability (std: {std_dance:.2f})")
    print("     → Popular songs don't all follow the same dance formula")
else:
    print(f"   • Low variation in danceability (std: {std_dance:.2f})")
    print("     → Popular songs tend to be similarly danceable")
```
