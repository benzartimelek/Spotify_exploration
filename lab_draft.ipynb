```python
#Splitting multi-genre entries and counting individual genres
genre_counts = {}
for genres in df['genre']:
    # Split by comma and clean whitespace
    genre_list = [g.strip() for g in genres.split(',')]
    for genre in genre_list:
        genre_counts[genre] = genre_counts.get(genre, 0) + 1
        
genre_df = pd.DataFrame(list(genre_counts.items()), columns=['Genre', 'Count'])
genre_df = genre_df.sort_values('Count', ascending=False)
```

```python
# Visualization of the Bar chart of genre distribution
plt.figure(figsize=(14, 6))
plt.bar(genre_df['Genre'], genre_df['Count'], color='steelblue', edgecolor='black')
plt.xlabel('Genre', fontsize=12, fontweight='bold')
plt.ylabel('Number of Songs', fontsize=12, fontweight='bold')
plt.title('Distribution of Songs Across Music Genres', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# The genre that has the most sonngs is:
print("Which genre has the most songs?")
top_genre = genre_df.iloc[0]
print(f"\n Genre with most songs: {top_genre['Genre']} ({top_genre['Count']} songs)")
```

```python
# The percentage of songs in top 3 genres are:
top_3_genres = genre_df.head(3)
total_genre_tags = genre_df['Count'].sum()
top_3_percentage = (top_3_genres['Count'].sum() / total_genre_tags) * 100

print("What percentage of songs are in the top 3 genres?")
print(f"\n Top 3 Genres:")
for idx, row in top_3_genres.iterrows():
    percentage = (row['Count'] / total_genre_tags) * 100
    print(f"   {row['Genre']}: {row['Count']} songs ({percentage:.1f}%)")
print(f"\n   Combined: {top_3_percentage:.1f}% of all genre tags")
```

**Task 2**

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Histogram
axes[0].hist(df['popularity'], bins=30, color='coral', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Popularity Score', fontsize=11, fontweight='bold')
axes[0].set_ylabel('Frequency', fontsize=11, fontweight='bold')
axes[0].set_title('Distribution of Song Popularity', fontsize=12, fontweight='bold')
axes[0].grid(axis='y', alpha=0.3)

# KDE plot
df['popularity'].plot(kind='kde', ax=axes[1], color='darkblue', linewidth=2)
axes[1].set_xlabel('Popularity Score', fontsize=11, fontweight='bold')
axes[1].set_ylabel('Density', fontsize=11, fontweight='bold')
axes[1].set_title('Popularity Distribution (KDE)', fontsize=12, fontweight='bold')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
#Calculate mean, median, and mode
popularity_mean = df['popularity'].mean()
popularity_median = df['popularity'].median()
popularity_mode = df['popularity'].mode()[0]

print(f"\n Popularity Statistics:")
print(f"- Mean:   {popularity_mean:.2f}")
print(f"- Median: {popularity_median:.2f}")
print(f"- Mode:   {popularity_mode}")
```

```python
# Box plot to identify outliers
plt.figure(figsize=(10, 6))
box_plot = plt.boxplot(df['popularity'], vert=True, patch_artist=True,
                        boxprops=dict(facecolor='lightgreen', color='darkgreen'),
                        medianprops=dict(color='red', linewidth=2),
                        whiskerprops=dict(color='darkgreen'),
                        capprops=dict(color='darkgreen'))
plt.ylabel('Popularity Score', fontsize=12, fontweight='bold')
plt.title('Box Plot of Song Popularity (Outlier Detection)', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#Typical popularity score range
q1 = df['popularity'].quantile(0.25)
q3 = df['popularity'].quantile(0.75)
iqr = q3 - q1

print("What is the typical popularity score range?")
print(f"\n Typical Popularity Range (IQR):")
print(f"   25th percentile (Q1): {q1}")
print(f"   75th percentile (Q3): {q3}")
print(f"   Interquartile Range:  {iqr}")
print(f"   Typical range: {q1} - {q3}")
```

```python
#Exceptional popularity
very_low = df[df['popularity'] < 20]
very_high = df[df['popularity'] > 80]

print(" Are there songs with exceptionally high or low popularity?")
print(f"\n Exceptional Cases:")
print(f"   Very low popularity (<20):  {len(very_low)} songs ({len(very_low)/len(df)*100:.1f}%)")
print(f"   Very high popularity (>80): {len(very_high)} songs ({len(very_high)/len(df)*100:.1f}%)")

if len(very_high) > 0:
    print(f"\n Most popular songs:")
    top_5 = df.nlargest(5, 'popularity')[['artist', 'song', 'popularity']]
    for idx, row in top_5.iterrows():
        print(f"   {row['artist']} - {row['song']} (Popularity: {row['popularity']})")
```

**Task 3**

```python
# Bar chart showing top 10 artists by number of songs.
top_artists = df['artist'].value_counts().head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_artists.index, top_artists.values, color='purple', edgecolor='black')
plt.xlabel('Artist', fontsize=12, fontweight='bold')
plt.ylabel('Number of Songs', fontsize=12, fontweight='bold')
plt.title('Top 10 Artists by Number of Songs', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Bar chart showing top 10 artists by average popularity
top_artists_popularity = df.groupby('artist')['popularity'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_artists_popularity.index, top_artists_popularity.values, color='blue', edgecolor='black')
plt.xlabel('Artist', fontsize=12, fontweight='bold')
plt.ylabel('Average Popularity', fontsize=12, fontweight='bold')
plt.title('Top 10 Artists by Average Popularity', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Does having more songs correlate with higher popularity?
artist_song_counts = df['artist'].value_counts()
artist_avg_popularity = df.groupby('artist')['popularity'].mean()
correlation_df = pd.DataFrame({
    'song_count': artist_song_counts,
    'avg_popularity': artist_avg_popularity
}).dropna()
correlation = correlation_df['song_count'].corr(correlation_df['avg_popularity'])
print("Does having more songs correlate with higher popularity?")
print(f'Correlation between number of songs and average popularity: {correlation:.2f}')
```

=> Nearly no linear relationship: The number of songs an artist releases has no big impact on how popular their music becomes on average. So number of songs is not a predictor og popularity.

```python
# Who are the most prolific artists in the dataset?
most_prolific_artists = artist_song_counts.head(5)
print("Who are the most prolific artists in the dataset?")
print("\n Most Prolific Artists:")  
for artist, count in most_prolific_artists.items():
    print(f"   {artist}: {count} songs")
```

**Task 4**

```python
# Calculate the percentage of songs with explicit content
explicit_count = df['explicit'].sum()
total_songs = len(df)   
explicit_percentage = (explicit_count / total_songs) * 100
print(f"\n Percentage of songs with explicit content: {explicit_percentage:.2f}%")
```

```python
# Pie chart or bar chart comparing explicit vs non-explicit songs.
labels = ['Explicit', 'Non-Explicit']
sizes = [explicit_count, total_songs - explicit_count]
colors = ['salmon', 'lightblue']
plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Explicit vs Non-Explicit Songs', fontsize=14, fontweight='bold')
plt.tight_layout()  
plt.show()
```

```python
# Box plot comparing popularity scores between explicit and non-explicit songs.
plt.figure(figsize=(6, 5))
sns.boxplot(x='explicit', y='popularity', data=df, hue='explicit', palette=['lightcoral', 'lightgreen'], medianprops=dict(color='red', linewidth=1), legend=False)
plt.xlabel('Explicit Content', fontsize=12, fontweight='bold')
plt.ylabel('Popularity Score', fontsize=12, fontweight='bold')
plt.title('Popularity Scores: Explicit vs Non-Explicit Songs', fontsize=14, fontweight='bold')
plt.xticks([0, 1], ['Non-Explicit', 'Explicit'])
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

- Explicit songs have a slightly higher median popularity

=> Explicit songs tend to be slightly more popular on average.


```python
# Do explicit songs tend to be more or less popular?
explicit_popularity = df[df['explicit'] == True]['popularity'].mean()
non_explicit_popularity = df[df['explicit'] == False]['popularity'].mean()

print("Do explicit songs tend to be more or less popular?")
print(f"\n Popularity Comparison:")
print(f"   Explicit songs average popularity:     {explicit_popularity:.2f}")
print(f"   Non-explicit songs average popularity: {non_explicit_popularity:.2f}")
print(f"   Difference: {explicit_popularity - non_explicit_popularity:.2f}")
```

=> Explicit songs tend to be slightly more popular than non-explicit songs, with an average popularity difference of about 2.2 points. However, the small gap suggests that explicit content alone is not a strong predictor of a songs success.
